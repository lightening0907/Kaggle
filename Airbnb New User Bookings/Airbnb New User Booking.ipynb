```python
def read_csv_zip(filename):
    z = zipfile.ZipFile(filename+'.zip','r')
    filename = filename.split('/')[-1]
    df = pd.read_csv(z.open(filename))
    return df
```

```python
def plot_bar(df, title,ylabel):
    """
    Helper function for plotting barplots.
    Color selection is made at random from a tuple of seabonrn colorsets
    """
    """
    p = (
        'Set2', 'Paired', 'colorblind', 'husl',
        'Set1', 'coolwarm', 'RdYlGn', 'spectral'
    )
    """
    color = sns.color_palette('hls', df.shape[1])
    fig = plt.figure(figsize=(15,15))
    
    bar   = df.plot(kind='barh',
                    title=title,
                    fontsize=16,
                    figsize=(12,8),
                    stacked=True,
                    width=1,
                    color=color,
    )
    plt.title(title,fontsize=20)
    
    plt.ylabel(ylabel,fontsize=20)
    # bar.figure.savefig(filename)

    plt.show()
```

```python
age_gb = age_gender.groupby(["age_bucket","country_destination"])["population_in_thousands"].sum().unstack()
```

```python
def DateTimeFeature(df,dateColumn,form):
    df[dateColumn+"DateIndex"] = pd.to_datetime(df[dateColumn],format=form)
    df[dateColumn+"Year"] = df[dateColumn+"DateIndex"].apply(lambda x: x.year)
    df[dateColumn+"Month"] = df[dateColumn+"DateIndex"].apply(lambda x: x.month)
    df[dateColumn+"Day"] = df[dateColumn+"DateIndex"].apply(lambda x: x.day)
    df[dateColumn+"WeekDay"] = df[dateColumn+"DateIndex"].apply(lambda x: x.weekday())
    #df[dateColumn+"Hour"] = df[dateColumn+"DateIndex"].apply(lambda x: x.hour) 
```

```python
categorical_features = ['gender','signup_method',
       'signup_flow', 'language', 'affiliate_channel',
       'affiliate_provider', 'first_affiliate_tracked', 'signup_app',
       'first_device_type', 'first_browser', 
       'date_account_createdMonth', 'date_account_createdWeekDay',
       'date_first_bookingMonth',
      'date_first_bookingWeekDay']
```

```python
continuous_features = [ 'age', 'date_account_createdYear',
       'date_first_bookingYear'] 
```

```python
train_user_na_removed = train_user.dropna(axis=0,how='any')
```

```python
def categorical_feature_preprocess(categorical_feature,dataset):
    list_categorical = []
    for col in categorical_feature:
        dummy = pd.get_dummies(dataset[col])
        list_categorical.append(dummy)
    dataset_categorical = pd.concat(list_categorical,axis=1)
    return dataset_categorical

train_categorical_na_removed = categorical_feature_preprocess(categorical_features,train_user_na_removed)
```

```python
train_data = pd.concat([train_user[continuous_features+['country_destination']],train_categorical],axis=1)
```

```python
from sklearn import preprocessing
```

```python
country_encoded = preprocessing.LabelEncoder()
country_encoded.fit(train_data["country_destination"])
train_data["country_encoded"] = country_encoded.transform(train_data["country_destination"])
```

```python
train_data_na_removed = train_data.dropna(axis=0,how='any')
```
